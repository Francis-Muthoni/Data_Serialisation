**To** **do** **list** **for** **a** **birthday** **party**


```python
def Welcome_banner():
    print("\nWelcome to todays to do list")
    print("1. Add Task")
    print("2. Mark Task as Completed")
    print("3. List Tasks")
    print("4. Remove Task")
    print("5. Quit the menu")

def add_task_to_be_accomplished(tasks):
    task_description = input("Enter the task description: ")
    tasks.append({"description": task_description, "completed": False})
    print(f'Task "{task_description}" added.')

def mark_task_complition_status(tasks):
    if tasks:
        list_tasks(tasks)
        try:
            task_index = int(input("Enter the task number to mark as completed: ")) - 1
            if 0 <= task_index < len(tasks):
                if not tasks[task_index]["completed"]:
                    tasks[task_index]["completed"] = True
                    print(f'Task "{tasks[task_index]["description"]}" marked as completed.')
                else:
                    print("This task is already marked as completed.")
            else:
                print("Invalid task number.")
        except ValueError:
            print("Please enter a valid number.")
    else:
        print("No tasks to mark as completed.")

def list_tasks(tasks):
    if tasks:
        print("\nTasks:")
        for index, task in enumerate(tasks, start=1):
            status = "Completed" if task["completed"] else "Pending"
            print(f'{index}. {task["description"]} - {status}')
    else:
        print("\nNo tasks to display.")

def remove_task(tasks):
    if tasks:
        list_tasks(tasks)
        try:
            task_index = int(input("Enter the task number to remove: ")) - 1
            if 0 <= task_index < len(tasks):
                removed_task = tasks.pop(task_index)
                print(f'Task "{removed_task["description"]}" removed.')
            else:
                print("Invalid task number.")
        except ValueError:
            print("Please enter a valid number.")
    else:
        print("No tasks to remove.")

def main():
    tasks = []
    while True:
        Welcome_banner()
        choice = input("Choose an option (1-5): ")
        if choice == '1':
            add_task_to_be_accomplished(tasks)
        elif choice == '2':
            mark_task_complition_status(tasks)
        elif choice == '3':
            list_tasks(tasks)
        elif choice == '4':
            remove_task(tasks)
        elif choice == '5':
            print("see you next time")
            break
        else:
            print("Invalid choice. Please choose a valid option.")

if __name__ == "__main__":
    main()
```
